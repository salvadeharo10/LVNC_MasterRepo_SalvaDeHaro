```python
import numpy as np
import pandas as pd
from sklearn.metrics import cohen_kappa_score, confusion_matrix
```

```python
file_path = "../Valoraciones_Cardiologos_Interobserver_review_AMIA2021.xlsx"
# Sheets' name
sheet_1 = "Valoración_Pepi"
sheet_2 = "Valoración_Daniel"
# We consider 0=not valid for assessment
df_1 = pd.read_excel(file_path, sheet_name=sheet_1, skiprows=2, usecols="C:O", index_col=0).replace({0:np.nan})
df_2 = pd.read_excel(file_path, sheet_name=sheet_2, skiprows=2, usecols="C:O", index_col=0).replace({0:np.nan})
```

## Considering scores as continious values

Inspired in http://dx.doi.org/10.21037/cdt.2017.03.12

```python
abs_diff = abs(scores_1-scores_2)
addition = scores_1+scores_2
rel_diff = abs_diff/(addition/2)
abs_diff
```

## Considering scores as categorical values

```python
scores_1 = scores_1.flatten()
scores_2 = scores_2.flatten()
mask = np.logical_and(~np.isnan(scores_1), ~np.isnan(scores_2))
scores_1, scores_2 = scores_1[mask], scores_2[mask]
```

```python
# https://stackoverflow.com/questions/11528150/inter-rater-agreement-in-python-cohens-kappa
cohen_kappa_score(scores_1*2, scores_2*2)
```

```python
(scores_1*2==scores_2*2).sum()/(len(scores_1))
```

## Weighted kappa

```python
cohen_kappa_score(scores_1*2, scores_2*2, sample_weight=abs_diff.flatten()[mask])
```

```python
df
```

```python
icc = pg.intraclass_corr(data=df, targets='slice', raters='rater', ratings='rating')
```
